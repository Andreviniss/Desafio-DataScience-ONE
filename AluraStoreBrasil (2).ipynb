### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#1. Análise do faturamento



```python
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

df_total = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

faturamento_por_loja = df_total.groupby('Loja')['Preço'].sum()

print("--- FATURAMENTO BRUTO POR LOJA ---")
tabela_estilizada = faturamento_por_loja.to_frame().style.format("R$ {:,.2f}")
tabela_estilizada

```

```python
faturamento_por_loja.plot(kind='bar', color='#3498db', figsize=(8,5))
plt.ylim(1300000, 1600000)
yticks_custom = np.arange(1300000, 1625000, 25000)
plt.yticks(yticks_custom)
plt.gca().set_yticklabels(['{:,.2f}M'.format(x/1000000) for x in yticks_custom])
plt.title('Diferença de Faturamento entre Lojas (Zoom: 1.3M - 1.6M)', fontsize=14)
plt.ylabel('Faturamento (em Milhões de R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
df_total['Data da Compra'] = pd.to_datetime(df_total['Data da Compra'], dayfirst=True)

vendas_por_loja_mes = df_total.set_index('Data da Compra').groupby('Loja').resample('M')['Preço'].sum().unstack(level=0)
vendas_por_loja_mes['Total Rede'] = vendas_por_loja_mes.sum(axis=1)

plt.figure(figsize=(12, 6))

for loja in [col for col in vendas_por_loja_mes.columns if col != 'Total Rede']:
    plt.plot(vendas_por_loja_mes.index, vendas_por_loja_mes[loja], marker='o', label=loja, alpha=0.7)

plt.plot(vendas_por_loja_mes.index, vendas_por_loja_mes['Total Rede'],
         marker='s', color='black', linewidth=3, linestyle='--', label='Total da Rede')

plt.title('Tendência de Faturamento Mensal: Lojas vs. Total da Rede', fontsize=14)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.legend(title='Legenda', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

# 2. Vendas por Categoria

```python
vendas_por_categoria = df_total.groupby(['Loja', 'Categoria do Produto'])['Preço'].sum()
vendas_por_categoria = vendas_por_categoria.sort_values(ascending=False)
print("--- FATURAMENTO POR CATEGORIA ---")
tabela_vendas = vendas_por_categoria.unstack()
print(tabela_vendas)
```

```python
vendas_detalhadas = df_total.groupby(['Loja', 'Categoria do Produto'])['Preço'].sum().unstack()

vendas_detalhadas.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

plt.title('Distribuição de Vendas por Categoria em Cada Loja', fontsize=14)
plt.ylabel('Faturamento Total (R$)', fontsize=12)
plt.xlabel('Lojas', fontsize=12)
plt.legend(title='Categorias', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

# 3. Média de Avaliação das Lojas

# 4. Produtos Mais e Menos Vendidos

```python
ranking_produtos_por_loja = df_total.groupby('Loja')['Produto'].value_counts(normalize=True)

print("--- TOP 3 PRODUTOS POR LOJA ---")
print(ranking_produtos_por_loja.groupby(level=0).head(3))

print("\n--- 3 PRODUTOS MENOS VENDIDOS POR LOJA ---")
print(ranking_produtos_por_loja.groupby(level=0).tail(3))
```

```python
plt.figure(figsize=(18, 12))

lojas = df_total['Loja'].unique()

for i, loja in enumerate(lojas):
    plt.subplot(2, 2, i + 1)

    df_loja = df_total[df_total['Loja'] == loja]

    contagem_produtos = df_loja['Produto'].value_counts()

    top_5 = contagem_produtos.head(5)
    bottom_5 = contagem_produtos.tail(5)

    df_top = top_5.to_frame(name='Vendas').assign(Categoria='Mais Vendidos')
    df_bottom = bottom_5.to_frame(name='Vendas').assign(Categoria='Menos Vendidos')


    if len(contagem_produtos) > 10:
        df_plot = pd.concat([df_top, df_bottom])
    else:
        df_plot = df_top


    sns.barplot(
        data=df_plot,
        x='Vendas',
        y=df_plot.index,
        hue='Categoria',
        dodge=False,
        palette={'Mais Vendidos': 'green', 'Menos Vendidos': 'red'}
    )

    plt.title(f'Top 5 Mais e Menos Vendidos - {loja}')
    plt.xlabel('Quantidade de Vendas')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

O objetivo é identificar padrões de vendas, eficiência logística e satisfação do cliente para recomendar estrategicamente qual unidade possui o menor potencial de retorno e, portanto, deve ser considerada para venda pelo Senhor João.

## 2. Análise de Dados

###Faturamento
Ao comparar o volume total de vendas, observamos uma disparidade clara. Enquanto a loja líder faturou **R\$ 1.534.509,12**, a **Loja 4** apresentou o menor desempenho financeiro, totalizando apenas **R\$ 1.384.497,58**. Isso representa uma diferença significativa de capital que impacta o fluxo de caixa do grupo.

##Despesas
Vejo que seria vital o conhecimento das despesas totais de cada loja para podermos nos basear no lucro líquido de cada loja mas como não tenho esses dados, irei me basear exclusivamente no faturamento bruto.

---

##Conclusão e Recomendação

Com base na análise quantitativa, recomenda-se que o Senhor João **venda a Loja 4**.

> **Justificativa:**
> Esta decisão é fundamentada no fato de a loja apresentar o **menor faturamento acumulado** entre todas as filiais analisadas. indicadores operacionais inferiores sugere que o esforço necessário para recuperar a rentabilidade desta unidade seria desproporcional comparado ao investimento nas lojas que já demonstram alto desempenho (como a Loja 1). A venda desta unidade liberará capital para fortalecer as operações mais lucrativas do grupo.

```python
resumo = df_total.groupby('Loja').agg({
    'Preço': 'sum',
    'Frete': 'mean',
    }).reset_index()

resumo.columns = ['Loja', 'Faturamento Total', 'Frete Médio']

pior_loja = resumo.sort_values('Faturamento Total').iloc[0]

print("--- DADOS PARA O RELATÓRIO ---")
print(resumo.style.format({'Faturamento Total': 'R$ {:,.2f}', 'Frete Médio': 'R$ {:.2f}'}).to_string())
print("\n" + "="*30)
print(f"SUGESTÃO DE VENDA: {pior_loja['Loja']}")
print(f"(Motivo: Menor faturamento, de R$ {pior_loja['Faturamento Total']:,.2f})")
print("="*30)

print(f"\n--- Top 3 Categorias da {pior_loja['Loja']} ---")
top_cat = df_total[df_total['Loja'] == pior_loja['Loja']]['Categoria do Produto'].value_counts().head(3)
print(top_cat)
```
